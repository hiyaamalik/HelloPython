### Variables

1. **No Declaration Required**
   Variables are created when a value is assigned.

   ```python
   x = 10
   ```

2. **Must Start with a Letter or Underscore**
   Valid:

   ```python
   name = "Python"
   _count = 5
   ```

   Invalid:

   ```python
   1num = 10   ‚ùå
   ```

3. **Cannot Start with a Number** üö´
   Variable names must not begin with digits.

4. **Only Letters, Numbers, and Underscores Allowed**
   Special characters like `@`, `$`, `%` are not allowed.

   ```python
   total_marks = 90
   ```

5. **Case-Sensitive** üî†

   ```python
   age = 20
   Age = 25   # Different variables
   ```

6. **No Reserved Keywords**
   Python keywords cannot be used as variable names.

   ```python
   class = 10   ‚ùå
   ```

7. **No Spaces Allowed**
   Use underscores instead.

   ```python
   first_name = "Alex"   ‚úÖ
   ```

8. **Meaningful Names Recommended** ‚úçÔ∏è

   ```python
   marks = 85   # Better than m
   ```

9. **Dynamic Typing Supported** üîÑ
   A variable can change its data type.

   ```python
   x = 10
   x = "Python"
   ```

### Level 1 Data Types

üî¢ 1. int (Integer)

Stores whole numbers

Can be positive or negative

````python
age = 50
count = -10
print(age)
print(type(age))
````

üî¢ 2. float

Stores decimal (fractional) numbers

````python
price = 99.99
pi = 3.14
print(price)
print(type(price))
````

Formatting the float using format command or f-string

````python
x = 3.14159
print(format(x, ".2f"))   # upto two decimal points

x = 3.14159
print(f"{x:.4f}")   # upto four decimal points
````

We can also round off, using the round function

````python
x = 3.14159

print(round(x, 2))   # Upto 2 places
print(round(x, 3))   # Upto 3 places
````

To truncate the decimal part, we simply type convert to int. It doesn't round off, instead removes the decimal part

````python
x = 3.9
print(int(x))
````

üìù 3. str (String)

### Type Conversion

Type conversion is the process of converting one data type into another data type in Python.

Why?

Sometimes, you need to perform operations between different data types, and Python requires them to match.

````python
x = "10"
y = int(x)

print(y)
print(type(y))
````

**Types of Conversion**

1. Implicit Conversion (Type Casting)

Done automatically by Python.

Usually happens when mixing types in operations.

````python
x = 5
y = 2.5
z = x + y
print(z)
````

2. Explicit Conversion (Type Casting)

Done manually using built-in functions.

Common Functions:

int() ‚Üí convert to integer

float() ‚Üí convert to float

str() ‚Üí convert to string

bool() ‚Üí convert to boolean

````python
x = "10"
y = int(x)
print(y + 5)
````
